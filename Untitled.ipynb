```python
def thresholding(img):
    
    thresh = threshold_otsu(img)
    
    binary = np.copy(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i,j] < thresh:
                binary[i,j] = 1
            else:
                binary[i,j] = 0
    return binary


def vertical_histogram(img):
    histo = np.zeros(img.shape[1])
    for i in range (img.shape[1]):
        histo[i] = np.sum(img[:,i])
    return histo

def horizontal_histogram(img):
    histo = np.zeros(img.shape[0])
    for i in range (img.shape[0]):
        histo[i] = np.sum(img[i,:])
    return histo
        
def get_width_line(line):
    v_histo = vertical_histogram(line)
    
    start =  0
    end = 0
    for i in range(len(v_histo)):
        if  v_histo[i] != 0:
            start  = i
            break            
    for i in range(len(v_histo)-1,0,-1):
        
        if  v_histo[i] != 0:
            end  = i
            break

    return start , end+1
        

    
def extract_lines(image):
    
    img = thresholding(image)
    
    lines = []
    
    h_hist = horizontal_histogram(img)
    
    flag = False
    
    start = 0
    end = 0
    
    new_image = np.ones(img.shape)
    
    for i in range(len(h_hist)):
        
        if  h_hist[i] != 0 and flag == False :
            flag =  True
            start = i
        
        if  h_hist[i] == 0 and h_hist[i-1] != 0  and flag == True :
            flag = False
            end = i
            lines.append((start,end))
    
    lines =  np.asarray(lines)
    
    
    LinesList = []
    
    for i in range(len(lines)):
        
        
        
        s,e = get_width_line(img[lines[i][0]:lines[i][1],:])
        
        image2[lines[i][0]:lines[i][1],s:e] = 0
        
        LinesList.append(image[lines[i][0]:lines[i][1],s:e])
    LinesList =  np.asarray(LinesList)
       
    return LinesList
        

    
    


image = rgb2gray(io.imread('Dataset/capr2.png'))

image2 =np.copy(image)

Lines =  extract_lines(image)


show_images([image2, image ],['',''])
```
